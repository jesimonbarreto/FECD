```python
n, p = 20, 0.15

x = np.arange(0,10)
r = binom.cdf(x, n, p)

r_values = list(range(10 + 1))
dist = [binom.pmf(r, n, p) for r in r_values ]
```

```python
plt.plot(x,r)
plt.show()
plt.bar(r_values, dist)
```

```python
posi_max = np.argmax(dist, axis=0)
print('k max é: {}. Posição max: {}'.format(dist[posi_max],posi_max))
print('Visualmente, a prosição onde a função se aproxima de 1 é entre 2 e 6')
print('Valor acumulado entre os valores estipulados {}'.format(sum([binom.pmf(i, n, p) for i in range(2,6)])))
print('Aproximação 0.95 = {}'.format(binom.ppf(0.95, n, p)))
print('Valor exato = {}'.format(binom.isf(1-0.95, n, p)))
print('E(X) tem valores altos proximos de 3')
```

```python
v_bin_rand = binom.rvs(n, p, loc=0, size=10000, random_state=None)
hist = plt.hist(v_bin_rand)
sum_v_rand = int(sum(hist[0][2:5]))
print('Foram preditos {} entre o intervalo 2 e 5 de 10000 aleatórios'.format(sum_v_rand))
print('Sim, os valores são parecidos entre a probabilidade e a geração de dados aleatórios')
```

# Questão 2

```python
mu = 0.73

x = np.arange(0,10)
r = poisson.cdf(x,mu)

r_values = list(range(10 + 1))
dist = [poisson.pmf(r, mu) for r in r_values ]

print('Valor k max é proximo de E(X) = λ = {}'.format(np.argmax(dist)))

plt.plot(x,r)
plt.show()
plt.bar(r_values, dist)
```

```python
mu = 10

x = np.arange(0,10)
r = poisson.cdf(x,mu)

r_values = list(range(10 + 1))
dist = [poisson.pmf(r, mu) for r in r_values ]

plt.plot(x,r)
plt.show()
plt.bar(r_values, dist)
```

```python
print('Valor k max é proximo de E(X) = λ = {}'.format(np.argmax(dist)))
print('Valor que se aproxima de 1 está entre 9 e 10')
print('Valor entre 9 e 10 é {}'.format(poisson.pmf(9, mu, loc=0)+poisson.pmf(10, mu, loc=0)))
```

```python
v_poi_rand = poisson.rvs(mu, loc=0, size=20000, random_state=None)
hist = plt.hist(v_poi_rand,20)
sum_v_rand = int(sum(hist[0][9:11]))
print('Foram preditos {} entre o intervalo 9 e 10 de 20000 aleatórios'.format(sum_v_rand))
print('Sim, os valores são parecidos entre a probabilidade e a geração de dados aleatórios')
```
